```python
from tqdm import tqdm
dataset = []
directory = "UrbanSound8K/audio/"
metadata = pd.read_csv("UrbanSound8K/metadata/UrbanSound8K.csv")

for index, row in tqdm(metadata.iterrows()):

  folder_num = row["fold"]
  file_name = row["slice_file_name"]
  label = row["class"]
  file_path = os.path.join(os.path.abspath(directory), "fold"+str(folder_num)+ "/", str(file_name) )
  audio, sampling_rate = librosa.load(file_path)
  mfcc_audio = librosa.feature.mfcc(audio, n_mfcc=13, hop_length = 512, sr = sampling_rate)

  if(len(mfcc_audio.T) == 173):
    dataset.append([mfcc_audio.T, label])
  else:
    continue
  # dataset.append([mfccs_scaled_features, label])
```
